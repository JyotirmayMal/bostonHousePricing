Preparing the Dataset

# CREATING LINEAR REGRESSION MODEL

```python
#independent features
X = dataset.iloc[:,:-1]
```

```python
#dependent features
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

TEST & TRAIN SPLIT

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 42)
```

STANDARDIZING THE DATASET

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

# MODEL TRAINING

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
#Check on which parameters, the model has been trained
regression.get_params()
```

# PREDICTION WITH TEST DATA

```python
reg_pred = regression.predict(X_test)
```

```python
#plot a scatter plot for the prediction
plt.scatter(y_test,reg_pred)
```

```python
#PREDICTION WITH RESIDUALS
residuals = y_test - reg_pred
sns.displot(residuals, kind = 'kde')
```

```python
plt.scatter(reg_pred,residuals)
```

# PERFORMANCE METRICES

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
print(mean_absolute_error(y_test,reg_pred))
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

```python
adjusted_r2_score = 1 - (1-score)*( (len(y_test) - 1)/(len(y_test) - X_test.shape[1] - 1) )
print(adjusted_r2_score)
```

PREDICTION OF NEW DATA

```python
d1 = scaler.transform(boston.data[0].reshape(1,-1))
```

```python
regression.predict(d1)
```
